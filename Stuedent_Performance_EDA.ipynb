```python
df['Total_Marks'] = df['writing score']+ df['reading score']+ df['math score']
df.head(5)
```

```python
df['Percentage'] = (df['Total_Marks']*100)/300
df.head(5)
```

```python
df[df['Percentage']<35]
```

The above calibration states that the student scoring an overall of less than 35% have not scored well in any of the three subjects.

```python
# Top 10 student's performance
df.sort_values(by='Percentage', ascending=False).head(10)
```

```python
sns.barplot(x='gender', y='Percentage', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='parental level of education', y='Percentage', data=df)
plt.show()
```

From the above graph we come to know that the students mark also depend upon the parental level of education. If the Parental level of education is high, those students have achieved better marks.

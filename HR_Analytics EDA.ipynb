```python
q1 = df['Age'].quantile(0.25)
q1
# Checking the age value which is at 25%
```

```python
q3 = df['Age'].quantile(0.75)
q3
# Checking the age value at 75%
```

```python
# Finding the IQR and the outliers of the age column
IQR = q3 - q1
lowl = q1 - 1.5*IQR
uppl = q3 + 1.5*IQR
print(IQR)
print(lowl)
print(uppl)
```

```python
sns.distplot(df['Age'])
plt.show()
```

```python
plt.hist(df.Age)
plt.show()
```

```python
# Treating the outliers for the Age feature
df = df[(df['Age'] > lowl) & (df['Age'] < uppl)]
df.shape
```

There are no outliers in the age column.

```python
# TO determin ethe skewness and kurtosis for age column
skew = df['Age'].skew()
print(skew)
kurt = df['Age'].kurt()
print(kurt)
```

--- Since the skewness is positive, which implies that the graph is rightly skewed.
--- The kurtosis is neagtive which implies that the data is of platykurtic natature.
